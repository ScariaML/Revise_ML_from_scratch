```python
import numpy as np 

class LR:
    def __init__(self, lr=0.01, iter = 1000):
        self.lr, self.iter, self.w, self.b = lr, iter, None, None
    def fit(self, X,y):
        m,n = X.shape
        self.w = np.zeros(n)
        self.b =0

        for _ in range(self.iter):
            y_pred = np.dot(X,self.w)+self.b

            dw = (1/m) * np.dot(X.T, (y_pred - y))
            db = (1/m) * np.sum(y_pred - y)
            self.w-=self.lr*dw
            self.b-=self.lr*db
    def pred(self, X):
        return np.dot(X,self.w)+self.b
            
```

```python
import pandas as pd 
df= pd.read_csv("2026-03-04_LinearRegression.csv")
df.head
X, y = df.drop("target", axis=1), df["target"].values

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import (mean_squared_error, r2_score)
Xtrain,Xtest, ytrain, ytest =train_test_split(X,y, test_size=0.2,random_state=7)
model = LR()
model.fit(Xtrain, ytrain)
ypred= model.pred(Xtest)
print(f"MSE and R2{ mean_squared_error(ytest,ypred),r2_score(ytest,ypred)}")
# print(np.isnan(ytest).sum())
```
