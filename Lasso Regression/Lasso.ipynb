```python
import numpy as np
class Lasso:
    def __init__ (self, lr=0.1, iter=10000, lambda_params=1.0):
        self.lr, self.iter, self.lambda_params= lr, iter, lambda_params
        self.w, self.b = None, None

    def fit(self, X, y):
        m,n = X.shape
        self.w = np.zeros(n)
        self.b = 0

        for _ in range(self.iter):
            y_hat = np.dot(X, self.w) +self.b
            # Subgradients of L1 penalty
            for j in range(n):
                if self.w[j]>0:
                    dw[j] = (1/m)* (np.dot(X[:,j].T,(y_hat-y))+(self.lambda_params/m)
                else:
                    dw[j] = (1/m)* (np.dot(X[:,j].T,(y_hat-y))-(self.lambda_params/m)
            db = (1/m)*np.sum(y_hat-y)

            self.w -=self.lr*dw
            self.b -=self.lr*db
    def predict(self, X):
    return np.dot(X, self.w) +self.b



              
```
