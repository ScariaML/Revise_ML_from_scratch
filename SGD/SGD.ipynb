```python
import numpy as np
import pandas as pd
class SGD:
    def __init__(self, lr =0.01, iter =30000):
        self.lr, self.iter, self.w, self.b = lr, iter, None, None
    
    def fit(self, X, y):
        m,n = X.shape
        self.w = np.zeros(n)
        self.b = 0

        for _ in range(self.iter):
            #shuffle the data to ensure randomness
            indices = np.random.permutation(m)
            Xshuffle = X[indices]
            yshuffle = y[indices]

            for i in range(m):
                xi = Xshuffle[i]
                yi = yshuffle[i]

                #predict for a simple sample
                yhat = np.dot(xi, self.w) +self.b

                #gradient for single sample
                error = yhat - yi
                dw = xi*error
                db = error
                self.w -=self.lr*dw
                self.b -=self.lr*db
    def predict(self, X):
        return np.dot(X, self.w)+self.b
```

```python
import pandas as pd
df = pd.read_csv("2026-03-13_SGDRegression.csv")
X, y =df.drop("target", axis=1).values, df["target"].values
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
X = Scaler.fit_transform(X)
print(X.shape, y.shape)
```

```python
model = SGD()
from sklearn.model_selection import train_test_split
from sklearn.metrics import (mean_squared_error, r2_score)
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
model.fit(Xtrain, ytrain)
ypred= model.predict(Xtest)
print(f"MSE and R2{ mean_squared_error(ytest,ypred),r2_score(ytest,ypred)}")
```
